# Problem: 
Predict canada's per capita income in year 2020. Given canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020.

```python
#import packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model
#Load and read dataset file
df = pd.read_csv('E:\Python\ML_Learn\Predict_Income_Model\canada_per_capita_income.csv')
df
```

```python
#Visualize dataset
plt.xlabel("Year")
plt.ylabel("per capita income (US$)")
plt.axis([1960,2020,1000,45000])
plt.scatter(df.year,df.loc[:,'per capita income (US$)'],color = "red",marker = "p")
```

```python
#Training process
reg = linear_model.LinearRegression()
X = df.drop("per capita income (US$)",axis = "columns")
y = df.loc[:,'per capita income (US$)']
reg.fit(X.values,y.values)
print(reg.coef_)
print(reg.intercept_)

```

```python
#Visualize fitting line
plt.xlabel("Year")
plt.ylabel("per capita income (US$)")
plt.axis([1960,2020,1000,45000])
plt.scatter(df.year,df.loc[:,'per capita income (US$)'],color = "red",marker = "p")
show_x = np.linspace(1960,2020)
show_y = reg.coef_ * show_x + reg.intercept_
show_y
plt.plot(show_x,show_y)
plt.show()
```

```python
#Predict per capital income in 2020
print(reg.predict([[2020]]))
```
