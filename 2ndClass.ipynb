```python
#Fibonacci while loop
def Fibonacci(n):
    if (n == 1 or n == 2):
        return 1
    first = 1
    second = 1
    #print(first)
    #print(second)
    while (n - 2 > 0):
        third = first + second 
        #print(third)
        first = second
        second = third
        n = n - 1
    return third
def Input(num):
    i = 1
    while (Fibonacci(i) <= num):
        print(str(Fibonacci(i))+' ', end ='')
        i = i + 1
Input(25)
```

```python
#Fibonacci with for loop
def Fibonacci(n):
    res = []
    if (n == 1):
        res = [0]
    if (n == 2):
        res = [0,1]
    else:
        res = [0,1]
    for i in range (1, n):
        res.append(res[i-1]+res[i])
    return res
Fibonacci(5)
```

```python
def Check (n):
    count = 0
    for i in range (1, int(n/2 + 1)):
        if n % i == 0:
            count = count + 1
    if count == 1:
        return True
    else:
        return False
def Output():
    res = []
    for i in range(0, 21):
        if Check(i) == True:
            res.append(i)
    return res
Output()
```
