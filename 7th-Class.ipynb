```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
```

```python
a = 2
b = 3
c = 5
noise = 0.3
x = np.linspace(-10,10,100)
y = a*pow(x,2) + b*x + c
x_train = x + noise*np.random.randn(1,100)
y_train = y + noise*np.random.randn(1,100)
plt.plot(x,y)
```

```python
plt.scatter(x_train,y_train)
```

```python
poly = PolynomialFeatures(degree = 2)
ones = np.ones((100,1))
a= np.reshape(x_train,(100,1))
train_x = poly.fit_transform(a)
train_x
reg = linear_model.LinearRegression()
reg.fit(train_x,np.reshape(y_train,(100,1)))
print(reg.coef_)
print(reg.intercept_)
```

```python
x_show = np.linspace(-10,10,100)
y_show = reg.intercept_ +reg.coef_[0][1] * x_show + reg.coef_[0][2]*pow(x_show,2)
plt.scatter(x_train,y_train)
plt.plot(x_show,y_show,color ="red")
```
