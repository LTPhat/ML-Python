```python
class Computer:
    def __init__(self):
        self.__maxprice = 900
        self.minprice = 500
    def sell(self):
        print("*********************************")
        print("Max selling price: {} ".format(self.__maxprice))
        print("Min selling price: {} ".format(self.minprice))
    def setMaxprice(self,maxprice):
        self.__maxprice = maxprice
c = Computer()
c.sell()
c.__maxprice = 1000  #Private variable does not change
c.sell()
c.setMaxprice(2000)
c.sell()
```

```python
#ENCAPSULATION (Tính bao hàm/đóng gói)
#Private variable in class can change only by method
class Car:
    def __init__(self, brand, color, fuel, max_speed, price):
        self.__brand = brand #Private
        self.color = color
        self.__fuel = fuel
        self.__maxspeed = max_speed  #Private
        self.price = price
    def show(self):
        print("************CAR INFORMATION*********************")
        print(self.__brand)
        print(self.color)
        print(self.__fuel)
        print(self.__maxspeed)
        print(self.price)
    def change_max_speed(self,speed):
        self.__maxspeed = speed
c = Car("Tesla", "White", "Electric",300,20000)
c.show()
c.color = "Red"
c.price = 1000
c.max_speed = 800
c.show()
c.change_max_speed(400)
c.show()
bmw = Car("BMW", "Blue", "Gas", 280,30000)
bmw.show()





        
```

```python
#POLYMORPHISM (Tính đa hình)

class Toyota:
    def dungxe(self):
        print("Toyota dừng xe để nạp điện")
    def nomay(self):
        print("Toyota nổ máy bằng hộp số tự động")
class Porsche:
    def dungxe(self):
        print("Porsche dừng xe để bơm xăng")
    def nomay(self):
        print("Porsche nổ máy bằng hộp số cơ")
def kiemtra_dung_xe(car): car.dungxe()

toyota = Toyota()
porsche = Porsche()
kiemtra_dung_xe(toyota)
kiemtra_dung_xe(porsche)
```
