# 1) Pyplot API and Object Oriented(OO) API

### Pyplot API

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),color = "blue",linestyle = "dashed",label ="sin x")
plt.plot(x,np.cos(x),color = "red",label = "cos x")
plt.title("Sin(x) and Cos(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.xlim([0,8])
plt.ylim([-1-0.1,1+0.1])
plt.legend()
plt.show()
```

### Object Oriented(OO) API

```python
%matplotlib inline
# 1) Prepare data
x = [1,2,3,4]
y = [11,22,33,44]
# 2) Setup plot 
fix, ax = plt.subplots(figsize = (5,5))
ax.plot(x,y)
ax.set(title = "A simple plot", xlabel = "x", ylabel = "y")
```

# 2) Common types of plot

### Line

```python
x = np.linspace(0,10,100)
#API Pyplot
plt.plot(x,x**3,color= "red")
```

```python
# OO API
fix,ax = plt.subplots()
ax.plot(x,x**3,color = 'red')
```

### Scatter

```python
plt.scatter(x,np.exp(x),color ="blue")
```

```python
#OO API
#Prepare new data
rng = np.random.RandomState(0)
x = rng.random(100)
y = rng.random(100)
colors = rng.rand(100)

fig,ax = plt.subplots()
ax.scatter(x,y,c = colors,cmap = "viridis")
```

### Bar

```python
#prepare data
soft_drink_price={"coke": 10, "pepsi": 12, "spite": 14, "heniken": 16}
fig,ax = plt.subplots()
ax.bar(soft_drink_price.keys(),soft_drink_price.values(),color = 'black',
       alpha = 0.5)
ax.set(title = "Soft drink price", xlabel = "Name", ylabel = "Price")
```
