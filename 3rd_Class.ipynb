```python
class Person:
    def dying_hair(self):
        self.hair = "red"
p1 = Person()
p1.eye = "black"
p1.hair = "black"
print(p1.eye)
print(p1.hair)
p1.dying_hair()
print(p1.hair)
```

```python
class Person:
    def dying_hair(self, color_hair):
        self.hair = color_hair
p1 = Person()
p1.eye = "black"
p1.hair = "black"
print(p1.eye)
print(p1.hair)
color_hair = "pink"
p1.dying_hair(color_hair)
print(p1.hair)
```

```python
class Person:
    def __init__(self, hair, eye, height, weight):
        self.hair = hair
        self.eye = eye
        self.height = height
        self.weight = weight
    def dying_hair(self, color):
        self.hair = color
p1 = Person("black", "brown", 172, 55)
print(p1.hair)
print(p1.eye)
print(p1.height)
print(p1.weight)
p1.dying_hair("green")
print(p1.hair)
```

```python
class Computer:
    def __init__(self):
        self.maxprice = 900
        self.minprice = 500
    def sell(self):
        print("*********************************")
        print("Max selling price: {} ".format(self.maxprice))
        print("Min selling price: {} ".format(self.minprice))
    def setMaxprice(self,maxprice):
        self.maxprice = maxprice
c = Computer()
c.sell()
c.maxprice = 1000
c.minprice = 4000
c.sell()
```
